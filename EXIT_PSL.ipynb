```python
##BIVARIATE ANALYSIS

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data creation for demonstration
data = pd.DataFrame({
    'player_id': range(1, 101),
    'name': ['Player'] * 100,
    'age': [25] * 100,
    'gender': ['M'] * 50 + ['F'] * 50,
    'country': ['CountryA'] * 50 + ['CountryB'] * 50,
    'height': [175] * 100,
    'weight': ['176lbs'] * 100,
    'ball_controlling_skills': [50] * 100,
    'body_reflexes': [60] * 100,
    'body_balance': [70] * 100,
    'jumping_skills': [80] * 100,
    'penalties_conversion_rate': [30] * 100,
    'mental_strength': [90] * 100,
    'goalkeeping_skills': [55] * 100,
    'defending_skills': [65] * 100,
    'passing_skills': [75] * 100,
    'dribbling_skills': [85] * 100,
    'shot_accuracy': [40] * 100,
    'body_strength_stamina': [95] * 100,
    'max_running_speed': [100] * 100,
    'strong_foot': ['Right'] * 100,
    'behaviour_rating': [4] * 100,
    'matches_played': [50] * 100,
    'fitness_rating': [80] * 100,
    'trophies_won': [1] * 100,
    'dedication_level': [5] * 100,
    'coaching': [3] * 100,
    'years_of_experience': [7] * 100,
    'no_of_disqualifications': [0] * 100,
    'selection': [1] * 100,
})

# Function to convert weight from string to float
def convert_weight(weight):
    if isinstance(weight, str):
        try:
            return float(weight.replace('lbs', '').strip())
        except:
            return None
    return float(weight)

# Apply the function to the weight column
data['weight'] = data['weight'].apply(convert_weight)

# Bivariate analysis
def bivariate_analysis(df, target):
    for column in df.columns:
        if column != target:
            print(f'Analysis of {column} vs {target}:')

            # If both columns are numerical
            if pd.api.types.is_numeric_dtype(df[column]) and pd.api.types.is_numeric_dtype(df[target]):
                sns.scatterplot(x=df[column], y=df[target])
                plt.title(f'Scatter plot of {column} vs {target}')
                plt.show()

            # If one column is numerical and the other is categorical
            elif pd.api.types.is_numeric_dtype(df[column]) and pd.api.types.is_string_dtype(df[target]):
                sns.boxplot(x=df[target], y=df[column])
                plt.title(f'Box plot of {column} vs {target}')
                plt.show()
            elif pd.api.types.is_string_dtype(df[column]) and pd.api.types.is_numeric_dtype(df[target]):
                sns.boxplot(x=df[column], y=df[target])
                plt.title(f'Box plot of {target} vs {column}')
                plt.show()

            # If both columns are categorical
            elif pd.api.types.is_string_dtype(df[column]) and pd.api.types.is_string_dtype(df[target]):
                contingency_table = pd.crosstab(df[column], df[target])
                sns.heatmap(contingency_table, annot=True, fmt="d")
                plt.title(f'Heatmap of {column} vs {target}')
                plt.show()

            print('\n')

# Specify the target column
target_column = 'selection'  # Replace with your actual target column name

# Run the bivariate analysis
bivariate_analysis(data, target_column)
```

```python
# Function to handle outliers using IQR
def handle_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    return data
```
