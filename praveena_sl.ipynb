```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/titanic_dataset.csv',index_col='PassengerId')
data.head()
```

```python
# Count missing values in each column
missing_values = data.isnull().sum()

# Print the number of missing values in each column
print(missing_values)
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
print(data.head())
```

```python
columns_with_outliers = ['Age', 'Fare', 'SibSp']

# IQR method
for column in columns_with_outliers:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers_iqr = data[(data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))]
    print(f"Number of outliers in {column} using IQR method: {len(outliers_iqr)}")
```

```python
for column in columns_with_outliers:
   Q1 = data[column].quantile(0.25)
Q3 = data[column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
print(data.head())
```
